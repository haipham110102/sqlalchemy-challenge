```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
conn = engine.connect()
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
first_row = session.query(Measurement).first()
first_row.__dict__
```

```python
# Find the most recent date in the data set.
first_row = session.query(Station).first()
first_row.__dict__
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Query All Records in the title table

measurement_data = pd.read_sql("SELECT * FROM Measurement", conn)
measurement_data.head()

```

```python
# Starting from the most recent data point in the database. 
latest_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
latest_date[0]
#earliest_date=session.query(Measurement.date).order_by(Measurement.date).first()
#earliest_date[0]
```

```python
# Calculate the date one year from the last date in data set.
#c=session.query(func.count(Measurement.date)).all()
#c[0][0]
latest_date =  dt.date(2017, 8, 23)
year_ago = latest_date - dt.timedelta(days=365)
#year_ago = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print("Year_ago: ", year_ago)
```

```python
# Perform a query to retrieve the data and precipitation scores
year_prcp = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >=year_ago).all()
#year_prcp = session.query(Measurement.date).filter(Measurement.date.between(year_ago, latest_date)).all()
year_prcp
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
year_prcp_df=pd.DataFrame(year_prcp, columns=['date','prcp'])
year_prcp_df.set_index('date')
# Sort the dataframe by date
year_prcp_df = year_prcp_df.sort_values("date")
year_prcp_df
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
year_prcp_df.plot(y='prcp', x='date', rot = 90)
plt.ylabel('Inch')
plt.title('Precipitation Scores')
plt.xticks(rotation="90")
plt.xlabel('Date')
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
year_prcp_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
stations =session.query(Measurement).group_by(Measurement.station).count()
print(stations)
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.

#SELECT measurement.station, count(measurement.station) AS count_1 
#FROM measurement GROUP BY measurement.station ORDER BY count(measurement.station) DESC
active_stations = session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()

active_stations_df=pd.DataFrame(active_stations, columns=['station','Number Stations'])

active_stations_df
```

```python
print(f"The most active station is {active_stations[0][0]}")
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.

min_max_avg_temp=session.query(func.min(Measurement.tobs),func.max(Measurement.tobs),func.round(func.avg(Measurement.tobs),2)).\
                   filter(Measurement.station == 'USC00519281').all()
print(f'The lowest temperature is {min_max_avg_temp[0][0]}')
print(f'The highest temperature is {min_max_avg_temp[0][1]}')
print(f'The average temperature is {min_max_avg_temp[0][2]}')
min_max_avg_temp = pd.DataFrame(min_max_avg_temp, columns=['min_temp','max_temp','avg_temp'])
min_max_avg_temp
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
#active_station_prcp = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date >= year_ago).filter(Measurement.station =='USC00519281').all()
active_station_tobs = session.query(Measurement.tobs).filter(Measurement.date >= year_ago, Measurement.station =='USC00519281').all()
active_station_tobs =pd.DataFrame(active_station_tobs, columns=['tobs'])
active_station_tobs
```

```python
active_station_tobs.plot.hist(bins=12)
plt.ylabel('Frequency')
plt.title('Temperature Observation Data (TOBS)')
plt.xticks(rotation="90")
plt.xlabel('Temperature')
plt.show()
```
